Doğrusal ilişkinin Gösterimi

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.lmplot(data= df, x="total_bill", y = "tip");
```

```python
sns.lmplot(data= df, x="total_bill", y = "tip", hue="smoker");
```

```python
sns.lmplot(data= df, x="total_bill", y = "tip", hue="smoker", col="time");
```

```python
sns.lmplot(data= df, x="total_bill", y = "tip", hue="smoker", col="day");
```

```python
sns.lmplot(data= df, x="total_bill", y = "tip", hue="smoker", col="time", row="sex");
```
