# Çizgi grafiği

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
df.timepoint.describe()
```

```python
df.signal.describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

Çizgi grafiği

```python
sns.lineplot(x = "timepoint", y="signal", data= df);
```

```python
sns.lineplot(x = "timepoint", y="signal", data= df, hue="event");
```

```python
sns.lineplot(x = "timepoint", y="signal", data= df, hue="event", style="event");
```

```python
sns.lineplot(x = "timepoint", y="signal", data= df, hue="event", style="event", markers=True, dashes=False);
```

```python
sns.lineplot(x = "timepoint", y="signal", data= df, hue="region", style="event", markers=True, dashes=False);
```
